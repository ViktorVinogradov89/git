Git - Это система контроля версий файлов
Главное в коммите — его атомарность. Другими словами,
он должен выполнять ровно одну задачу
Команда git status не выводит файлы,
которые добавлены в репозиторий и не содержат изменений

Функции
- Создание репозиториев - хранилищ файлов
- Отслеживание изменений в файлах в локальном репо и серверном репо
- Склейка (сливание) изменений из 1 ветки в 2 ветку

- Инструментарий
Служебные файлы, добавляемые операционной системой — например,
.DS_Store в MacOS
Конфигурационные и временные файлы редакторов — например,
.idea или .vscode

Временные файлы
- Логи — в них содержится полезная информация для отладки,
которая собирается во время запуска и работы приложения
- Кеши — файлы, которые нужны для ускорения разных процессов

Артефакты
- Результаты сборки проекта — например, после компиляции
или сборки фронтенда
Зависимости, которые устанавливаются во время разработки — например,
 node_modules или vendor
- Результаты выполнения тестов — например, информация о покрытии
кода тестами
- После установки Git нужно настроить. Для своей работы ему 
важно знать ваше имя и почту. 
Эти данные подставляются в историю изменений. Только так можно узнать,
кто и что сделал в проекте:
- Выполняется из любой директории
  - C:\Users\USER>git config --global user.name 'своё Имя'
  - C:\Users\USER>git config --global user.email 'свой mail'

Основные команды:
- mkdir hexlet  git Создаем новый проект
- touch создание файла
- cd hexlet  git Переходим в созданную директорию
- git init (Выполняем инициализацию)создание репозитория
- git clone взять репозиторий из внешнего хранилища на локальное
- git add добавить файл для будущего коммита (в индекс / staging area)
- git add . добавляет все измененные файлы в текущей папке 
или подпапках
- git add -i показывает измененные куски файлов
git commit -am коммитет все изменения
- git commit -m добавляет файлы для коммита в локальный commit
- git commit --amend после этой команды откроется редактор,
ожидающий ввода описания коммита
- git push добавляет из локального commit во внешний репозиторий
- git pull применяет изменения из внешнего репозитория
в локальный репозиторий
- git log посмотреть изменения.
- git log --oneline Показывает сокращенный вывод
    Тут все коммиты с полным дифом
    Промотать вперед — кнопка f, промотать назад — b
    Выйти из режима просмотра — q
- git reset удаление коммита(можно удалять коммиты, отменять 
их без удаления, восстанавливать файлы из истории и так далее)
- git reset --hard HEAD~ (--hard означает полное удаление,
 HEAD~ означает «один коммит от последнего коммита».) 
Если бы мы хотели удалить два последних коммита,
 то могли бы написать HEAD~2
- reset --soft HEAD~1 (--soft отменяет коммит,
 но оставляет изменения в индексе)
- git revert создаёт новый коммит, 
который отменяет изменения предыдущего.
- git show посмотреть хэш коммита
- git blame <путь до файла> эта команда выводит файл
 и рядом с каждой строчкой показывает того, 
 кто ее менял и в каком коммите
- git grep 'любое слово' ищет слово во всех файлах проекта 
- git checkout HEAD~ Переключится на предыдущий коммит.
- git checkout 'название ветки' вернуться в ветку
- git checkout 'номер хэша' гит переходит в определенную версию
- git checkout <хеш коммита> перемещаться по истории,
 загружая в рабочую директорию состояние кода на момент 
 выполнения любого коммита
- git status можно посмотерть статус репозитория
- git remote add origin git@github.com:<ИМЯ НА ГИТХАБЕ>/hexlet-git.git
  создание репозитория в гите
- git branch - узнать место нахождения
- git push -u origin main
- git clone git@github.com:<ИМЯ НА ГИТХАБЕ>/hexlet-git.git -
 клонирование репозитория
- git pull --rebase скачивает из внешнего репозитория новые коммиты
 и добавляет их в локальный репозиторий
 - sed -i '2i\Вторая строка' имя_файла.txt
(sed - это потоковый редактор.
-i означает "встроенное редактирование" (изменения применяются
 непосредственно к файлу).
`2i\` - это инструкция sed, которая вставляет строку перед
 строкой с номером 2 (т.е. второй строкой). `\` означает, 
 что следующая строка - это вставляемый текст.
Вторая строка - это текст, который нужно добавить. 
Замените его на нужную строку.
имя_файла.txt - замените на имя вашего файла.)
- nano 'имя файла' открыть файл в терминале (ctrl+ O, enter, Ctrl+X)
- rm 'имя файла' удаление файла
- git clean -fd удалить все файлы в дериктории
- cat 'имя файла' вывести информацию в файле в терминал
- git mv 'старое имя' 'новое имя' переименовать файл
- git restore 'имя файла'  восстанавливаем удалённый файл
- git restore --staged 'имя файла' отменяет индексацию, 
но не отменяет изменения,сделанные в файле
- git reset --hard HEAD отменяет все изменения,
 сделанные в рабочей директории и индексе Git
- git diff  показывает разницу между тем, что было и что стало
- git diff --staged выведет все изменения, 
сделанные в рабочей директории которые были добавлены в индекс
- git restore для отмены изменений в таких файлах используется команда
причем Git сам напоминает об этом при проверке статуса
- git rm --cached my_file.txt позволяет удалить файл из индекса
 (области подготовки) Git, но при этом сохранить файл в вашей рабочей директории
- git stash не удаляет файлы. Они попадают в специальное место внутри директории
 .git на временное хранение.
- git stash pop после выполнения всех нужных изменений в 
чистой рабочей директории можно вернуть спрятанные изменения 
- echo 'Hello, Hexlet!' > README.md

Создание ssh ключей:
  - ssh-keygen -t ed25519  -C "your_email@example.com"
- Дальше будет несколько вопросов. На все вопросы нужно нажимать Enter.
- Запуск агента ssh, который следит за ключами
  - eval "$(ssh-agent -s)"
- Добавления нового ssh-ключа в агент
  - ssh-add ~/.ssh/id_ed25519

Интеграция с Git:
- Выведите содержимое файла ~/.ssh/id_ed25519.pub и скопируйте его:
  - cat ~/.ssh/id_ed25519.pub
- Перейдем в раздел и вставим ключ
  - https://github.com/settings/keys
- Проверьте, что подключение работает
  - ssh -T git@github.com


Алгоритм решения проблем в компьютер сайнс(Open source):
* если код открыт и я могу его прочитать, модифицировать и расширять
 - это Open source (Язык програмирования, линтер, программы, CICD(deploy))
1. ишем в чем ошибка(код ошибки, инструкция к которой происходит ошибка)
 в располощении где находится ошибка. Поиск файла
2. Находим в фале инструкцию, которая вызывает код ошибки.
3. Зайти в Github, найти тот же файл, программу, через глобальный поиск,
 или напрямую через репозиторий, в котором находится линтер, рограмма и тд.
 Весь Open source берется с гита.
4. Воспроизведения ошибки, проведение тех операций которые 
способствуют его проявлению
5. У официального репозитория показательный фактор является большое
 количество звезд (если новый ппроект божет быть меньше)
6. В issues пользователь заявляет о проблеме и там решают её. 
Каждый пул реквест прикреплен к issues для решения той или иной проблемы.
(Заводим пул только если есть навыки решения проблемы)
